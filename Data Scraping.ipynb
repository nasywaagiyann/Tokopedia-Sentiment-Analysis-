```python
import pandas as pd  # Mengimpor library pandas untuk manipulasi data tabular

# Menginstal library google-play-scraper untuk mengambil data ulasan dari Google Play Store
!pip install google-play-scraper

# Mengimpor fungsi yang diperlukan dari library google_play_scraper
from google_play_scraper import Sort, reviews_all

# Mengambil semua ulasan aplikasi Tokopedia dari Google Play Store
reviews = reviews_all(
    'com.tokopedia.tkpd',     # ID aplikasi Tokopedia di Play Store
    sleep_milliseconds=0,     # Waktu jeda antar permintaan (0 berarti tanpa jeda)
    lang='id',                # Bahasa ulasan yang diambil (Indonesia)
    country='id',             # Negara asal ulasan (Indonesia)
    sort=Sort.NEWEST,         # Mengambil ulasan terbaru terlebih dahulu
)

# Mengonversi hasil ulasan menjadi DataFrame menggunakan pandas
df = pd.DataFrame(reviews)

# Mengambil maksimal 100.000 ulasan teratas (jika tersedia)
df = df.head(100000)

# Menyimpan data ulasan ke dalam file CSV dengan encoding UTF-8
df.to_csv('ulasan_tokopedia_new.csv', index=False, encoding='utf-8')

# Menampilkan jumlah ulasan yang berhasil diambil dan disimpan
print(f"Scraped {len(df)} reviews and saved to ulasan_tokopedia_new.csv")
```
